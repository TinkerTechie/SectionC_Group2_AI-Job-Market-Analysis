# 2. Data Cleaning and Structuring
## Raw Data Loading

```python
import pandas as pd

# Load the raw dataset extracted in the previous notebook
df = pd.read_csv('../data/raw_dataset.csv')
print(f"Initial shape: {df.shape}")
```

## Handling Missing Values
Identifying missing data points and applying appropriate imputation techniques.

```python
# Check for nulls in Location
missing_locs = df['Location'].isnull().sum()
print(f"Missing values in Location: {missing_locs}")

# Since the missing locations represent a specific subset, we will impute using the mode
# which maintains the structural integrity of the geographic distribution.
mode_location = df['Location'].mode()[0]
df['Location'] = df['Location'].fillna(mode_location)

print(f"Missing values after imputation: {df['Location'].isnull().sum()}")
```

## Standardizing Categorical Variables
Resolving inconsistencies in textual data by mapping variations of categorical features.

```python
# Standardize 'Required Education' terminology
education_mapping = {
    'bachelors': 'Bachelor\'s Degree',
    'Masters': 'Master\'s Degree'
}

df['Required Education'] = df['Required Education'].replace(education_mapping)

print("Standardized Required Education categories:")
print(df['Required Education'].value_counts().head())
```

## Numerical Data Transformation
Extracting pure numerical values from string representations and casting them to appropriate types.

```python
# 1. Clean 'Median Salary (USD)' by removing the '$' and casting back to float
df['Median Salary (USD)'] = df['Median Salary (USD)'].str.replace('$', '', regex=False).astype(float)

# 2. Clean 'Experience Required (Years)' by removing ' yrs' and casting back to int
df['Experience Required (Years)'] = df['Experience Required (Years)'].str.replace(' yrs', '', regex=False).astype(int)

# Verify final data types
print("\n--- Cleaned Data Types ---")
print(df[['Median Salary (USD)', 'Experience Required (Years)']].dtypes)

# Save the final, perfectly structured dataset
df.to_csv('../data/cleaned_dataset.csv', index=False)
print("\nPipeline complete. Cleaned dataset saved as 'cleaned_dataset.csv'")
display(df.head())
```
